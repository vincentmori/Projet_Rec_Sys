```python
import unicodedata
import re 
def remove_accents(text):
    # 1. Normalisation en forme NFD (Normalisation Form D - Decomposed)
    #    'é' devient 'e' + accent aigu
    normalized_text = unicodedata.normalize('NFD', text)
    
    # 2. Utiliser les expressions régulières pour supprimer les caractères non-ASCII (les accents)
    #    \p{Mn} correspond à tous les 'Mark, nonspacing' (les accents, trémas, etc.)
    cleaned_text = re.sub(r'[\u0300-\u036f]', '', normalized_text)
    
    return cleaned_text

print(remove_accents(city))
```
